```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frequency = [1.126-0.16,1.126-0.08,1.126-0.05,1.126-0.03,1.126,1.126+0.03,1.126+0.05,1.126+0.08,1.126+0.16]  # 频率范围, 表二,MHZ
amplitude = [1.8,4.8,5.2,5.6,6,5.4,4.2,2.4,1.4]  # 振幅,V
amplitude = [i/2 for i in amplitude]
plt.rcParams['font.family'] = 'Microsoft YaHei'
plt.figure(figsize=(6, 4))
# plt.subplot(1, 2, 1)
plt.plot(frequency, amplitude)
plt.scatter(frequency, amplitude, color='red')  # 显示每个点
for i, txt in enumerate(amplitude):
    plt.text(frequency[i], amplitude[i], f'({frequency[i]:.2f}, {amplitude[i]:.2f})', fontsize=8, ha='left', va='bottom')
plt.title('幅度-频率曲线')
plt.xlabel('频率 / MHZ')
plt.ylabel('电压 / V')
plt.show()
```

```python
distance = [10, 13, 16, 19, 22, 25, 28, 31]  # 距离范围, cm,表四
voltage = [2.4,3.6,4.6,5.4,5.0,4.6,4.2,3.6]  # 电压,V
plt.figure(figsize=(6, 4))
plt.plot(distance, voltage)
plt.scatter(distance, voltage, color='blue')  # 显示每个点
for i, txt in enumerate(voltage):
    plt.text(distance[i], voltage[i], f'({distance[i]:.2f}, {voltage[i]:.2f})', fontsize=8, ha='left', va='bottom')
plt.title('电压-距离曲线')
plt.xlabel('距离 / cm')
plt.ylabel('电压 / V')
plt.show()
```
